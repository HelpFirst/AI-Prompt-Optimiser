```python
# !pip install pycaret
import pandas as pd
from pycaret.classification import *
```

```python
eval_datapath = "news.csv"
text_column = "Title"
target_column = "topic" 
df = pd.read_csv(eval_datapath, encoding='ISO-8859-1', usecols=[text_column, target_column])
df.sample(5)
```

```python
nlp = setup(data=df, target=target_column, text_features=[text_column], session_id=1)
```

```python
compare_models()
```

```python
model = create_model('ridge')
```

```python
evaluate_model(model)
```

```python
model2 = create_model('et')
```

```python
evaluate_model(model2)
```
